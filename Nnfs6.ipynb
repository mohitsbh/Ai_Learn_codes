```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create training data
x = np.linspace(-5, 5, 50)     # 50 points between -5 and 5
y = np.sin(x)                   # target output

# 2. Choose number of RBF centers
numCenters = 10

# Select random centers from training data
randIndex = np.random.choice(len(x), numCenters, replace=False)
centers = x[randIndex]

# 3. Define Gaussian Radial Basis Function
def rbf(x, c, sigma):
    return np.exp(-((x - c)**2) / (2 * sigma**2))

# 4. Compute RBF activations (Phi matrix)
sigma = 1                                 # spread parameter
Phi = np.zeros((len(x), numCenters))

for i in range(numCenters):
    Phi[:, i] = rbf(x, centers[i], sigma)

# 5. Compute weights using pseudo-inverse
W = np.linalg.pinv(Phi) @ y

# 6. Predict outputs
y_pred = Phi @ W

# 7. Plot actual vs predicted
plt.figure()
plt.plot(x, y, 'b', linewidth=2, label='Actual sin(x)')           # Actual sin(x)
plt.plot(x, y_pred, 'r', linewidth=2, label='RBFN Output')      # RBFN Output
plt.legend()
plt.title("RBF Network Approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# 8. Display result
print("RBF Network Training Complete")
```
