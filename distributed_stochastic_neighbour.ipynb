```python
from sklearn.manifold import TSNE
import numpy as np

# X: your high-dimensional data, shape (n_samples, n_features)
X = np.array([[0, 1, 2],
              [1, 0, 3],
              [2, 1, 0],
              [3, 2, 1]])

tsne = TSNE(n_components=2, random_state=42, perplexity=3)
X_embedded = tsne.fit_transform(X)
X_embedded.shape # (n_samples, 2)
print(X_embedded)
```
