```python

import numpy as np
import matplotlib.pyplot as plt
# Triangular MF
def trimf(x, a, b, c):
 return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)
# Trapezoidal MF
def trapmf(x, a, b, c, d):
 return np.maximum(np.minimum(np.minimum((x - a) / (b - a), 1), (d - x) / (d - c)), 0)
# Gaussian MF
def gaussmf(x, c, sigma):
 return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))
# Sigmoid MF
def sigmf(x, a, c):
 return 1 / (1 + np.exp(-a * (x - c)))
# Generalized Bell MF
def gbellmf(x, a, b, c):
 return 1 / (1 + np.abs((x - c) / a) ** (2 * b))
# Example usage
x = np.linspace(0, 10, 500)
fig, axs = plt.subplots(5, 1, figsize=(6, 12)) # 5 rows, 1 column
# Triangular
axs[0].plot(x, trimf(x, 2, 5, 8), 'b')
axs[0].set_title("Triangular MF")
axs[0].set_ylabel("Membership degree")
axs[0].grid(True)
# Trapezoidal
axs[1].plot(x, trapmf(x, 2, 4, 6, 8), 'g')
axs[1].set_title("Trapezoidal MF")
axs[1].set_ylabel("Membership degree")
axs[1].grid(True)
# Gaussian
axs[2].plot(x, gaussmf(x, 5, 1), 'r')
axs[2].set_title("Gaussian MF")
axs[2].set_ylabel("Membership degree")
axs[2].grid(True)
# Sigmoid
axs[3].plot(x, sigmf(x, 2, 5), 'm')
axs[3].set_title("Sigmoid MF")
axs[3].set_ylabel("Membership degree")
axs[3].grid(True)
# Generalized Bell
axs[4].plot(x, gbellmf(x, 2, 4, 5), 'c')
axs[4].set_title("Generalized Bell MF")
axs[4].set_xlabel("x")
axs[4].set_ylabel("Membership degree")
axs[4].grid(True)
plt.tight_layout()
plt.show()
```
