```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the universe of discourse
x = np.linspace(0, 10, 1000)

# Define a fuzzy membership function (Triangular)
membership = fuzz.trimf(x, [2, 5, 8])  # Triangular MF between 2, 5, and 8

# Defuzzification Methods
centroid = fuzz.defuzz(x, membership, 'centroid')   # 1. Centroid (Center of Gravity)
bisector = fuzz.defuzz(x, membership, 'bisector')   # 2. Bisector
mom = fuzz.defuzz(x, membership, 'mom')             # 3. Mean of Maximum
som = fuzz.defuzz(x, membership, 'som')             # 4. Smallest of Maximum
lom = fuzz.defuzz(x, membership, 'lom')             # 5. Largest of Maximum

# Plot the membership function
plt.figure(figsize=(8, 5))
plt.plot(x, membership, 'b', linewidth=1.5, label='Membership Function')

# Mark defuzzified results
for val, method in zip(
    [centroid, bisector, mom, som, lom],
    ['Centroid', 'Bisector', 'MOM', 'SOM', 'LOM']
):
    plt.axvline(x=val, linestyle='--', linewidth=1.2, label=method)

plt.title("Defuzzification Methods")
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Degree")
plt.legend()
plt.show()

# Print crisp values
print("Centroid:", round(centroid, 2))
print("Bisector:", round(bisector, 2))
print("MOM:", round(mom, 2))
print("SOM:", round(som, 2))
print("LOM:", round(lom, 2))
```
