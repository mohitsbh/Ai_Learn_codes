```python
import numpy as np

# Parameters
alpha = 0.1  # learning rate
gamma = 0.6  # discount factor
epsilon = 0.1  # exploration rate

# States and actions
Q = np.zeros((6, 2))

# Environment rewards and transitions
rewards = np.array([-1, -1, -1, -1, 10, -1])
actions = [0, 1]

def choose_action(state):
    if np.random.uniform(0, 1) < epsilon:
        return np.random.choice(actions)  # explore
    else:
        return np.argmax(Q[state])  # exploit learned values

# Q-learning loop
for episode in range(1000):
    state = 0
    while state != 4:  # goal state
        action = choose_action(state)
        next_state = (state + 1) if action == 1 else max(state - 1, 0)
        reward = rewards[next_state]

        # Q-table update
        Q[state, action] = Q[state, action] + alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])
        state = next_state

print("team q-table:")
print(Q)
```
