```python
import pandas as pd

url=("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

Data=pd.read_csv(url)
Data
```

```python
Data.info()
```

```python
Data.describe()
```

```python
Data.Age.isnull().sum()
```

```python
Age=Data.Age.mean()
```

```python
Data.Age=Data.Age.fillna(Age)
```

```python
Data.Age.isnull().sum()
```

```python
Data.Embarked.isnull().sum()
```

```python
Data.Embarked.mode()[0]
```

```python
Data.Embarked=Data.Embarked.fillna(Data["Embarked"].mode()[0])
```

```python
Data["Embarked"].isnull().sum()
```

```python
Data_first=Data[Data.Pclass==1].head()
print("passengersData_first \n",Data_first)
```
